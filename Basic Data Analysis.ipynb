```python
# import pandas
import pandas as pd
```

```python
#Read the dataset and store it in a dataframe
df = pd.read_excel('Absenteeism.xls')
```

```python
# settings to display all columns
pd.set_option("display.max_columns", None)
#print the shape of the dataframe
print("Shape of the dataset", df.shape)
# print the first 5 instances and 
df.head()
```

```python
#printing number of rows and number of columns
print('Number of rows: ', df.shape[0])
print('Number of columns: ', df.shape[1])
```

```python
#printing the column names
print(df.columns)
```

```python
#printing the column names in the form of a list
col_list = [] # this is an empty list
for x in df.columns:
    col_list.append(x)
print(col_list)
col_list
```

```python
from matplotlib import pyplot as plt 
import seaborn as sns
```

```python
print(df.columns.tolist())
```

```python
#Let's see how the transortation expense varies with distance from residence to work
plt.figure(figsize = (8,8))
plt.scatter(x = 'Distance from Residence to Work', y = 'Transportation expense', data = df)
plt.xlabel('Distance from residence to work')
plt.ylabel('Transportation Expense')
plt.title('Scatterplot of transportation expense against distance from residence to work');
```

# Box and Whisker Plots

```python
import random
colorlist = ['red', 'yellow', 'orange', 'purple', 'blue']
sns.boxplot(x = 'Distance from Residence to Work', data = df, color = random.choice(colorlist));
```

```python
sns.boxplot(y = 'Distance from Residence to Work', x = 'Disciplinary failure', data = df);
```

```python
for i in df.columns:
    plt.figure()
    sns.boxplot(x=i, data = df)
```

```python
for i in df.columns:
    plt.figure(figsize=(6,6))
    sns.boxplot(y=i, x = 'Disciplinary failure', data = df)
```
